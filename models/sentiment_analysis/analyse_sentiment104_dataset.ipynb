```python
import pandas as pd
import matplotlib.pyplot as plt
from langdetect import detect
```

## Understand the data
Load the original Sentiment140 data file and check:
1. The distribution of the value of labels
2. Print the text that is classified as non English by langdetect

```python
df = pd.read_csv('data/tweets-with-sentiment.csv')

# keep only the 'target' and 'text' columns
df = df[['target', 'text']]

# check the value distribution of the 'target' column
target_distribution = df['target'].value_counts()
print(target_distribution)

def is_english(text):
  language = detect(text)
  if language != 'en':
    print(text, language)

# validate whether all tweets are in English
df['text'].apply(is_english)
```

## Plot the distribution of word count of tweets after preprocess
In this step, we will load the preprocessed data file generated by sentiment_analysis_w_lstm.ipynb

```python
df = pd.read_csv('data/pre.csv')

# Check the value distribution of the 'target' column
target_distribution = df['target'].value_counts()
print(target_distribution)

# Step 2: Filter Data
target_0 = df[df['target'] == 0]['text_len']
target_4 = df[df['target'] == 1]['text_len']

# Step 3: Plotting
plt.figure(figsize=(10, 6))

# Plot histogram for target 4
plt.hist(target_4, bins=20, alpha=0.5, color='red', label='Positive')

# Plot histogram for target 0
plt.hist(target_0, bins=20, alpha=0.5, color='blue', label='Negative')


plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.title('Distribution of Word Count for Tweets with Positive and Negative Sentiment')
plt.legend()
plt.grid(True)
plt.show()
```
